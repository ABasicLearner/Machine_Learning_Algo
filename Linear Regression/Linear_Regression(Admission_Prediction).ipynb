```python
import pickle
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV, LinearRegression
from sklearn.model_selection import train_test_split
from ydata_profiling import ProfileReport
```

```python
df
```

GRE Score: This is the score of the applicant on the Graduate Record Examination (GRE). The GRE is a standardized test that is used for graduate school admissions.

TOEFL Score: This is the score of the applicant on the Test of English as a Foreign Language (TOEFL). The TOEFL is a standardized test that is used to measure English proficiency for international students.

University Rating: This is a subjective rating of the university that the applicant is applying to. The rating is on a scale of 1 to 5, with 1 being the lowest and 5 being the highest.

SOP: This is the applicant's Statement of Purpose. The SOP is a personal statement that describes the applicant's academic and professional goals.

LOR: This is the applicant's Letter(s) of Recommendation. The LOR is a letter from a professor or other academic advisor who can attest to the applicant's academic abilities and potential.

CGPA: This is the applicant's cumulative grade point average (GPA). The GPA is a measure of the applicant's academic performance in college.

Research: This is a binary variable that indicates whether or not the applicant has research experience.

Chance of Admit: This is the predicted probability of the applicant being admitted to the graduate program.

```python
#Adjusted R square
def adj_r2(x, y):
    r2 = lr.score(x, y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(x_test, y_test)
```
